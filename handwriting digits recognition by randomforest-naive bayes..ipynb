```python
from keras.datasets import mnist
```

```python
# Load data

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
# Data attributes

print("train_images dimentions: ", train_images.ndim)

print("train_images shape: ", train_images.shape)

print("train_images type: ", train_images.dtype)
```

```python
print('number of train_data:',len(train_images))
print ('number of test_data:',len(test_images))
```

```python
train_images[0].shape
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
imgplot = plt.imshow(train_images[4000])

plt.gray()

plt.show()
```

```python
X_train = train_images.reshape(60000, 784)
X_test = test_images.reshape(10000, 784)
```

```python
#Classification Algorithm
from sklearn.naive_bayes import GaussianNB
classifier_Naive=GaussianNB()
classifier_Naive.fit(X_train,train_labels)
out_Navie=classifier_Naive.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay

print(confusion_matrix(test_labels,out_Navie))

print(' ')

cm_display=ConfusionMatrixDisplay(confusion_matrix(test_labels,out_Navie))

cm_display.plot()

plt.show()
```

```python
from sklearn.metrics import classification_report, accuracy_score

print('Accuracy= ' + str(round(accuracy_score(test_labels, out_Navie)*100)))
```

```python
from sklearn.metrics import classification_report, accuracy_score

print(classification_report(test_labels,out_Navie))
```

```python
#Classification Algorithm
from sklearn.ensemble import RandomForestClassifier
classifier_randomforest=RandomForestClassifier()
classifier_randomforest.fit(X_train,train_labels)
out_randomforest=classifier_randomforest.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay

print(confusion_matrix(test_labels,out_randomforest))

print(' ')

cm_display=ConfusionMatrixDisplay(confusion_matrix(test_labels,out_randomforest))

cm_display.plot()

plt.show()
```

```python
from sklearn.metrics import classification_report, accuracy_score
print('Accuracy= ' + str(round(accuracy_score(test_labels, out_randomforest)*100)))
```

```python
print(classification_report(test_labels,out_randomforest))
```
